## Feature Definitions

id :- It is the unique numeric number assigned to each house being sold.

date :- It is the date on which the house was sold out.

price:- It is the price of house which we have to predict so this is our target variable and aprat from it are our features.

bedrooms :- It determines number of bedrooms in a house.

bathrooms :- It determines number of bathrooms in a bedroom of a house. Decimal values refer to cases where a bathroom is not complete (e.g. A complete room is one with a toilet, sink and shower)

sqft_living :- It is the measurement variable which determines the measurement of house in square foot.

sqft_lot : It is also the measurement variable which determines square foot of the lot.

floors: It determines total floors means levels of house.

waterfront : This feature determines whether a house has a view to waterfront 0 means no 1 means yes.

view : This feature determines whether a house has been viewed or not 0 means no 1 means yes.

condition : It determines the overall condition of a house on a scale of 1 to 5.

grade : It determines the overall grade given to the housing unit, based on King County grading system on a scale of 1 to 11

sqft_above : It determines square footage of house apart from basement.

sqft_basement : It determines square footage of the basement of the house.

yr_built : It detrmines the date of building of the house.

yr_renovated : It detrmines year of renovation of house.

zipcode : It determines the zipcode of the location of the house.

lat : It determines the latitude of the location of the house.

long : It determines the longitude of the location of the house.

sqft_living15 : The square footage of interior housing living space for the nearest 15 neighbors 

sqft_lot15 : The square footage of the land lots of the nearest 15 neighbors

# Data Preperation

```python
housedatadf = housedata.drop(["id", "date"],axis=1)
#removing sqft_lot, sqft_lot15 and yr_built only reduced the accuracy.
housedatadf.head(5)
```

```python
#remove outliers
housedatadf = housedatadf[housedatadf.bedrooms!=33]
housedatadf = housedatadf[housedatadf.bedrooms!=11]


housedatadf = housedatadf[housedatadf.sqft_living <= 13000]

housedatadf = housedatadf[housedatadf.bathrooms != 7.5]

housedatadf = housedatadf[housedatadf.floors < 3.0]



#modify values

#separate renovated houses and unrenovated houses
housedatadf.loc[housedatadf['yr_renovated'] == 0, 'yr_renovated'] = 0
housedatadf.loc[housedatadf['yr_renovated'] > 0 , 'yr_renovated'] = 1

#separate houses with and without basement, retaining the size of the basement as it still 
#affect house prize
housedatadf['basement'] = np.nan
housedatadf.loc[housedatadf['sqft_basement'] == 0, 'basement'] = 0
housedatadf.loc[housedatadf['sqft_basement'] > 0, 'basement'] = 1

#separate condition into 2 groups
housedatadf.loc[housedatadf['condition'] <= 2, 'condition'] = 0
housedatadf.loc[housedatadf['condition'] >= 3 , 'condition'] = 1

housedatadf.head(5)
```

```python
corr = housedatadf.corr(method = 'pearson')
corr.sort_values(["price"], ascending = False, inplace = True)
print(corr.price)
```

# Training Model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

y = housedatadf["price"].values
X = housedatadf.drop(["price"],axis=1).values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=129)
```

## Train Model with Algorithm: Linear Regression 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr
```

## Train Model with Algorithm: Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(max_depth=5 ,random_state=129)
tree.fit(X_train,y_train)
tree
```

## Train Model with Algorithm: Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)

poly_model=LinearRegression()
poly_model.fit(X_train_poly,y_train)
```

# Score and Evaluate model

```python
# model evaluation for training set
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

## Score and Evaluate model: Linear Regression

```python
# model evaluation for training set
y_hat= lr.predict(X_test)
ty_hat = lr.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train, ty_hat)))
r2 = r2_score(y_train, ty_hat)
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
rmse = (np.sqrt(mean_squared_error(y_test, y_hat)))
# r-squared score of the model
r2 = r2_score(y_test, y_hat)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
